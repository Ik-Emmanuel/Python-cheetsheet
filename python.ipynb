# Dictionaries

```
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "price": 12898
}
```

```
my_keys = car.keys()
print(list(my_keys))
my_values = car.values()
print(list(my_values))
items = car.items()
print(items)
```

```
#get dictionary values
value1 = car.get('brand',None)
print(value1)
value2 = car.get('Website',None)
print(value2)
```

```
#adding a key to a dict if it doesnt exist
print(car)
value1 = car.setdefault('brand', "My own Ford")
print(car)
value2 = car.setdefault('Website', "WWW.cars.com")
print(car)
print(value1)
print(value2)
```

### Default Dict

```
from collections import defaultdict
```

```
def_dict = defaultdict(str)
print(def_dict)
print(def_dict['name'])
print(def_dict)
```

```
#Used to create a default dict with given values if you ask for a key it creates it 
my_dict = defaultdict(lambda: "default_value")
print(dict(my_dict))
print(my_dict['users'])
print(my_dict['houses'])
print(dict(my_dict))
```

```
#creating dictionaries from collection
a = [['name', 'Jenny'],['age', '35']]
b = [('user', 'mike'), ('age', 4)]
new_dict1 = dict(a)
new_dict2 = dict(b)
print(new_dict1)
print(new_dict2)
print(new_dict1['name'])
```

```
keys = ['a', 'b', 'c']
values = [1, 2, 3]
my_dict = dict(zip(keys, values))
print(my_dict)
my_dict.update({'c': 4, 'd':10})
print(my_dict)
my_dict.pop('c')
print(my_dict)
```

## Dictionary Comprehension

```
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd':2, 'd':2}
dicts = {k for k, v in my_dict.items() if v == 2}
new_dicts = {k:v for k, v in my_dict.items() if v == 2}
print(dicts)
print(type(dicts))
print(new_dicts)
print(type(new_dicts))
```
